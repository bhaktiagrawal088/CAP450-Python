```python
meat_to_animal = {
'bacon': 'pig',
'pulled pork': 'pig',
'pastrami': 'cow',
'corned beef': 'cow',
'honey ham': 'pig',
'nova lox': 'salmon'
}

def meat_2_animal(series):
    if series['food'] == 'bacon':
        return 'pig'
    elif series['food'] == 'pulled pork':
        return 'pig'
    elif series['food'] == 'pastrami':
        return 'cow'
    elif series['food'] == 'corned beef':
        return 'cow'
    elif series['food'] == 'honey ham':
        return 'pig'
    else :
        return 'salmon'
data =  pd.DataFrame({'food' : ['bacon' , 'pulled pork' , 'pastrami','corned beef' , 'honey ham']})
data['animal'] = data['food'].map(str.lower).map(meat_to_animal)
data
```
